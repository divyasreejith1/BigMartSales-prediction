```python
#Data preprocessing
```

```python
#filling of the missing values of column 'Item_weight' with mean
```

```python
train_data1['Item_Weight']=train_data1['Item_Weight'].fillna(train_data1['Item_Weight'].mean())
```

```python
#filling the missing values of column'Outlet_Size' with mode since it is a categorical column
```

```python
train_data1['Outlet_Size']=train_data1['Outlet_Size'].fillna(train_data1['Outlet_Size'].mode()[0])
```

```python
train_data1.isna().sum()
```

```python
#Log Transformation
```

```python
train_data1['Item_Outlet_Sales']=np.log(1+train_data1['Item_Outlet_Sales'])
```

```python
#Encoding of Categorical columns 
```

```python
#One hot Encoding
```

```python
train_data1=pd.get_dummies(train_data1)
```

```python
train_data1
```

```python
#splitting the features and target
```

```python
X=train_data1.drop('Item_Outlet_Sales',axis=1)
y=train_data1['Item_Outlet_Sales']
```

```python
X
```

```python
X.describe()
```

```python
#splitting the data into train and test
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
#Building model
```

# Linear Regression model

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
lr=LinearRegression()
model1=lr.fit(X_train,y_train)
pred1=model1.predict(X_test)
```

```python
pred1
```

```python
pd.DataFrame(pred1)
```

```python
#Mean Squared Error and R2 score
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print('Mean squared error is: ',mean_squared_error(y_test,pred1))
print('Rsquared value is: ',r2_score(y_test,pred1))
```

# Lasso Regression Model

```python
ls=Lasso(alpha=0.009)
model2=ls.fit(X_train,y_train)
pred2=model2.predict(X_test)
```

```python
pred2
```

```python
#Mean Squared Error and R2 score
```

```python
print('Mean squared error is: ',mean_squared_error(y_test,pred2))
print('Rsquared value is: ',r2_score(y_test,pred2))
```

# Ridge Regression

```python
rid_reg=Ridge(alpha=0.009)
model3=rid_reg.fit(X_train,y_train)
pred3=model3.predict(X_test)
```

```python
#Mean Squared Error and R2 score
```

```python
print('Mean squared error is: ',mean_squared_error(y_test,pred3))
print('Rsquared value is: ',r2_score(y_test,pred3))
```

```python
import base64
```

```python
from IPython.display import HTML


def create_download_link( test_data, title = "Download CSV file", filename = "data.csv"):  
    csv = test_data.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

create_download_link(test_data)
```
